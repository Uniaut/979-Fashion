```python
import pandas as pd

# annotation to dataframe
anno_df = pd.DataFrame(json_data['annotations'])

anno_df = anno_df.explode('attribute_ids')

anno_df.head()

# cross table of category and attribute
cat_attr_ct = pd.crosstab(anno_df['category_id'], anno_df['attribute_ids'], normalize='index')

# fill no attribute index with 0
cat_attr_ct = cat_attr_ct.reindex(range(46), fill_value=0)
# fill no category index with 0
cat_attr_ct = cat_attr_ct.reindex(range(341), axis=1, fill_value=0)

# plot croos table as pyplot
cat_attr_np = cat_attr_ct.to_numpy()
# x axis: category 0~45
# y axis: attribute 0~340
plt.figure(figsize=(40, 40))
plt.yticks(np.arange(0, 45, 5))
plt.xticks(np.arange(0, 340, 5))
plt.xlabel('attr')
plt.ylabel('cat')
plt.title('cross table of category and attribute')
# plt.grid(True, which='both')
plt.autoscale(enable=True, axis='x', tight=True)

plt.imshow(np.power(cat_attr_np, 0.2), cmap='hot', interpolation='nearest')
```
